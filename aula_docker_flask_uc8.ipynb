# 1. Por que estudar Docker?

Imagine a seguinte situação:

Um desenvolvedor cria uma API Flask no computador dele. Na máquina dele tudo funciona, porque ele tem:

- Python instalado;
- Flask instalado;
- versão correta das bibliotecas;
- arquivos organizados;
- variáveis de ambiente configuradas;
- sistema operacional compatível.

Mas quando outro colega tenta rodar o mesmo projeto, acontece algum erro:

> "Professor, na minha máquina não funciona."

Isso é muito comum em desenvolvimento de software.

O Docker ajuda a resolver esse problema porque permite empacotar a aplicação junto com tudo que ela precisa para funcionar.

Em vez de depender da configuração de cada computador, criamos um ambiente padronizado.

# 2. O que é Docker?

Docker é uma ferramenta usada para criar, empacotar e executar aplicações em ambientes isolados chamados **containers**.

Um container contém tudo que a aplicação precisa para rodar:

- código da aplicação;
- bibliotecas;
- dependências;
- configurações;
- comandos de execução;
- ambiente base.

De forma simples:

> Docker permite colocar uma aplicação dentro de uma "caixa" padronizada para que ela rode de forma parecida em diferentes ambientes.

Essa "caixa" é o container.

# 3. Relação entre Docker e Cloud Computing

Docker é muito usado em ambientes de nuvem.

Em empresas, é comum que aplicações sejam empacotadas em containers e executadas em:

- servidores;
- máquinas virtuais;
- serviços de containers;
- plataformas cloud;
- Kubernetes;
- serviços como Azure, AWS, Google Cloud, Render, Railway, entre outros.

A ideia é:

> Criar a aplicação uma vez, empacotar em uma imagem e executar em diferentes ambientes.

Isso facilita:

- implantação;
- escalabilidade;
- manutenção;
- padronização;
- redução de erros de ambiente;
- automação de deploy.

# 4. Conceitos principais

Antes de usar Docker, precisamos entender três conceitos fundamentais:

1. **Dockerfile**
2. **Imagem Docker**
3. **Container**

Esses três conceitos estão relacionados.

## 4.1 Dockerfile(receita do bolo)

O `Dockerfile` é um arquivo de texto que funciona como uma **receita**.

Ele diz ao Docker como montar o ambiente da aplicação.

Exemplo:

```dockerfile
FROM python:3.12-slim

WORKDIR /app

COPY requirements.txt .

RUN pip install --no-cache-dir -r requirements.txt

COPY . .

EXPOSE 5000

CMD ["python", "app.py"]
```

Esse arquivo responde perguntas como:

- Qual imagem base será usada?
- Qual pasta será usada dentro do container?
- Quais arquivos serão copiados?
- Quais bibliotecas serão instaladas?
- Qual porta a aplicação usa?
- Qual comando inicia a aplicação?

Resumo:

> Dockerfile é a receita para criar uma imagem Docker.

## 4.2 Imagem Docker(imagem do bolo)

A imagem Docker é o pacote gerado a partir do Dockerfile.

Ela contém a aplicação e tudo que ela precisa para funcionar.

Você pode pensar na imagem como um "modelo pronto" ou uma "foto congelada" do ambiente.

Para criar uma imagem, usamos:

```bash
docker build -t minha-api-flask .
```

Resumo:

> A imagem é o resultado do Dockerfile.

## 4.3 Container(comendo o bolo)

O container é uma imagem em execução.

Quando rodamos uma imagem, criamos um container.

Exemplo:

```bash
docker run -p 5000:5000 minha-api-flask
```

Resumo:

> Container é a aplicação rodando dentro do ambiente criado pela imagem.

# 12. Criando a API Flask

Agora vamos criar o arquivo `app.py`.

Este será o código da nossa API.

Observe que estamos usando uma lista em memória chamada `tarefas`.

Isso significa que os dados existem apenas enquanto a aplicação está rodando. Se a aplicação for reiniciada, os dados criados durante a execução podem ser perdidos.

Em aplicações reais, normalmente usamos um banco de dados.

````python
from flask import Flask, jsonify, request

app = Flask(__name__)

tarefas = [
    {
        "id": 1,
        "titulo": "Estudar Docker",
        "descricao": "Entender imagem, container e Dockerfile",
        "concluida": False
    },
    {
        "id": 2,
        "titulo": "Criar API Flask",
        "descricao": "Criar endpoints GET, POST, PUT e DELETE",
        "concluida": False
    }
]


@app.route("/", methods=["GET"])
def home():
    return jsonify({
        "mensagem": "API de Tarefas funcionando!"
    })


@app.route("/health", methods=["GET"])
def health():
    return jsonify({
        "status": "ok",
        "mensagem": "Aplicação em execução"
    })


@app.route("/tarefas", methods=["GET"])
def listar_tarefas():
    return jsonify(tarefas)


@app.route("/tarefas/<int:id_tarefa>", methods=["GET"])
def buscar_tarefa(id_tarefa):
    for tarefa in tarefas:
        if tarefa["id"] == id_tarefa:
            return jsonify(tarefa)

    return jsonify({"erro": "Tarefa não encontrada"}), 404


@app.route("/tarefas", methods=["POST"])
def criar_tarefa():
    dados = request.get_json()

    if not dados:
        return jsonify({"erro": "Nenhum dado foi enviado"}), 400

    if "titulo" not in dados or "descricao" not in dados:
        return jsonify({
            "erro": "Os campos titulo e descricao são obrigatórios"
        }), 400

    nova_tarefa = {
        "id": len(tarefas) + 1,
        "titulo": dados["titulo"],
        "descricao": dados["descricao"],
        "concluida": dados.get("concluida", False)
    }

    tarefas.append(nova_tarefa)

    return jsonify(nova_tarefa), 201


@app.route("/tarefas/<int:id_tarefa>", methods=["PUT"])
def atualizar_tarefa(id_tarefa):
    dados = request.get_json()

    if not dados:
        return jsonify({"erro": "Nenhum dado foi enviado"}), 400

    for tarefa in tarefas:
        if tarefa["id"] == id_tarefa:
            tarefa["titulo"] = dados.get("titulo", tarefa["titulo"])
            tarefa["descricao"] = dados.get("descricao", tarefa["descricao"])
            tarefa["concluida"] = dados.get("concluida", tarefa["concluida"])

            return jsonify(tarefa)

    return jsonify({"erro": "Tarefa não encontrada"}), 404


@app.route("/tarefas/<int:id_tarefa>", methods=["DELETE"])
def deletar_tarefa(id_tarefa):
    for tarefa in tarefas:
        if tarefa["id"] == id_tarefa:
            tarefas.remove(tarefa)

            return jsonify({
                "mensagem": "Tarefa removida com sucesso"
            })

    return jsonify({"erro": "Tarefa não encontrada"}), 404


if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)
````

# 13. Observação importante sobre `host="0.0.0.0"`

No final do código usamos:

```python
app.run(host="0.0.0.0", port=5000, debug=True)
```

Isso é importante quando a aplicação roda dentro de um container.

## Diferença entre 127.0.0.1 e 0.0.0.0

### 127.0.0.1

Significa:

> Aceitar conexões apenas do próprio ambiente local.

Dentro de um container, isso pode fazer a API ficar acessível apenas internamente no container.

### 0.0.0.0

Significa:

> Aceitar conexões externas ao ambiente onde a aplicação está rodando.

Por isso, quando usamos Docker, normalmente configuramos Flask com:

```python
host="0.0.0.0"
```

Assim, conseguimos acessar a API pelo navegador do computador.

# 16. Dockerfile comentado

Abaixo está a mesma versão do Dockerfile, agora com comentários explicando cada linha.

```dockerfile
# Define a imagem base com Python instalado
FROM python:3.12-slim

# Define a pasta de trabalho dentro do container
WORKDIR /app

# Copia o arquivo de dependências para dentro do container
COPY requirements.txt .

# Instala as bibliotecas listadas no requirements.txt
RUN pip install --no-cache-dir -r requirements.txt

# Copia todos os arquivos do projeto para dentro do container
COPY . .

# Informa que a aplicação usa a porta 5000 dentro do container
EXPOSE 5000

# Comando executado quando o container iniciar
CMD ["python", "app.py"]
```

# 17. Explicando o Dockerfile linha por linha

## FROM

```dockerfile
FROM python:3.12-slim
```

Define a imagem base.

Estamos dizendo:

> Comece a imagem usando uma versão leve do Python 3.12.

## WORKDIR

```dockerfile
WORKDIR /app
```

Define a pasta de trabalho dentro do container.

É parecido com fazer:

```bash
cd /app
```

## COPY requirements.txt .

```dockerfile
COPY requirements.txt .
```

Copia o arquivo `requirements.txt` da máquina para dentro do container.

O ponto `.` representa a pasta atual dentro do container.

## RUN

```dockerfile
RUN pip install --no-cache-dir -r requirements.txt
```

Executa um comando durante a criação da imagem.

Aqui estamos instalando o Flask.

## COPY . .

```dockerfile
COPY . .
```

Copia todos os arquivos do projeto para dentro do container.

## EXPOSE

```dockerfile
EXPOSE 5000
```

Documenta que a aplicação usa a porta 5000.

Importante:

> EXPOSE não publica a porta automaticamente.

Para publicar, usamos `-p` no `docker run`.

## CMD

```dockerfile
CMD ["python", "app.py"]
```

Define o comando que será executado quando o container iniciar.

# 18. RUN x CMD

Essa diferença é muito importante.

| Instrução | Quando acontece? | Exemplo |
|---|---|---|
| RUN | Durante a criação da imagem | Instalar bibliotecas |
| CMD | Quando o container inicia | Rodar a aplicação |

Exemplo:

```dockerfile
RUN pip install -r requirements.txt
```

Esse comando instala dependências enquanto a imagem está sendo criada.

```dockerfile
CMD ["python", "app.py"]
```

Esse comando roda quando o container é iniciado.

# 22. Entendendo o mapeamento de portas

O comando:

```powershell
docker run -p 5000:5000 minha-api-flask
```

usa o parâmetro `-p`.

Ele segue este formato:

```text
porta_do_computador:porta_do_container
```

Então:

```text
5000:5000
```

significa:

> Tudo que chegar na porta 5000 do computador será enviado para a porta 5000 do container.

Outro exemplo:

```powershell
docker run -p 8080:5000 minha-api-flask
```

Nesse caso:

- dentro do container, a aplicação roda na porta 5000;
- no computador, acessamos pela porta 8080.

Acesso:

```text
http://localhost:8080
```

Isso é útil quando a porta 5000 já está ocupada.

Quando usamos:

```powershell

```

o terminal fica preso mostrando os logs da aplicação.

Para rodar em segundo plano, usamos `-d`:


O `-d` significa **detached mode**.

Ou seja:

> O container continua rodando, mas o terminal fica livre.

```markdown
## Descrição

Esta API permite cadastrar, listar, atualizar e remover tarefas. O projeto foi desenvolvido com Flask e executado dentro de um container Docker.

- Python
- Flask
- Docker
- JSON

| Método | Endpoint | Descrição |
|---|---|---|

pip install -r requirements.txt

## Como rodar o container
```

# 39. Conclusão

Nesta aula, você aprendeu a transformar uma API Flask em uma aplicação containerizada.

```text
Criar API Flask
        ↓
```
